### Input: An array A of n elements
### Output: Sorted array in ascending order
Steps of Merge Sort Algorithm

1Ô∏è‚É£ If the array has one or zero elements, return it (already sorted).

2Ô∏è‚É£ Divide the array into two equal halves.

3Ô∏è‚É£ Recursively apply Merge Sort on the left half.

4Ô∏è‚É£ Recursively apply Merge Sort on the right half.

5Ô∏è‚É£ Merge the two sorted halves into a single sorted array.

6Ô∏è‚É£ Return the merged sorted array.

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result

ans= merge_sort([38, 27, 43, 3, 9, 82, 10])
print(ans)
```
