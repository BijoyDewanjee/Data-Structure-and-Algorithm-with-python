‚ö° Algorithm: Quick Sort
Input:

An array A of n elements

Output:

The array sorted in ascending order

üîπ Steps of Quick Sort Algorithm

1Ô∏è‚É£ Choose a pivot element from the array
    (usually first, last, middle, or random element)

2Ô∏è‚É£ Partition the array such that:

All elements smaller than or equal to pivot are placed on the left

All elements greater than pivot are placed on the right

3Ô∏è‚É£ The pivot element is now in its correct position.

4Ô∏è‚É£ Recursively apply Quick Sort on the left subarray.

5Ô∏è‚É£ Recursively apply Quick Sort on the right subarray.

6Ô∏è‚É£ Stop when the subarray has one or zero elements.


```python
def quick_sort(arr, low, high):
    if low < high:
        p = partition(arr, low, high)
        quick_sort(arr, low, p - 1)
        quick_sort(arr, p + 1, high)

def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1

    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

arr = [3, 6, 8, 10, 1, 2, 1]
quick_sort(arr, 0, len(arr) - 1)
print(arr)
```
