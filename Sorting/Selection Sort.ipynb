üîç Algorithm: Selection Sort
## Input: An array A of n elements
Output: Sorted array in ascending order
Steps:

1Ô∏è‚É£ Start from the first element of the array.

2Ô∏è‚É£ Assume the first element is the minimum.

3Ô∏è‚É£ Compare this minimum element with the remaining elements of the array.

4Ô∏è‚É£ If a smaller element is found, update the minimum.

5Ô∏è‚É£ After completing the inner loop, swap the minimum element with the first element of the unsorted part.

6Ô∏è‚É£ Move the boundary of the sorted part one position forward.

7Ô∏è‚É£ Repeat steps 2‚Äì6 until the entire array is sorted.

```python
def selection_sort(arr):
    n = len(arr)

    for i in range(n - 1):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j

        arr[i], arr[min_index] = arr[min_index], arr[i]

    return arr

result = selection_sort([5, 1, 4, 2, 8])
print(result)
```
