# Insertion Sort

1️⃣ Start from the second element of the array (index = 1), because the first element is already sorted.

2️⃣ Store the current element in a variable called key.

3️⃣ Compare the key with the elements before it (left side).

4️⃣ Shift all elements that are greater than key one position to the right.

5️⃣ Insert the key into its correct position.

6️⃣ Repeat steps 2–5 for all remaining elements of the array.

7️⃣ Stop when the entire array becomes sorted.

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key

    return arr

result = insertion_sort([12, 11, 13, 5, 6])
print(result)
```

```python
result2 = insertion_sort([8, 3, 5, 2, 6])
print(result2)
```
